```python

import os

directory = "/home/user/Documents/tp"
dir_exist = True
yes_or_no = ''
count = 0
nb0 = 0

def printSuccess(message):
    print("\n---\n")
    print('-----' + message + '-----')
    print("\n---\n")

def printError(message):
    print("\n---\n")
    print('-----' + message + '-----')
    print("\n---\n")

def isExistDirectory(dirname):
    try:
        os.chdir(dirname)
    except:
        return False
    return True

def question1():

    global dir_name1     
    global dir_name11
    global dir_name12
    global dir_name2     
    global dir_name21
    global dir_name22

    dir_name1 = input("Entrez le nom du repertoire1: ")
    dir_name11 = dir_name1 + '.1'
    dir_name12 = dir_name1 + '.2'

    dir_name2 = input("Entrez le nom du repertoire2: ")
    dir_name21 = dir_name2 + '.1'
    dir_name22 = dir_name2 + '.2'

    """
        avant de creer les repertoires, tester dabord si ils existe ou pas>>> 
        
    """
    
    if not(isExistDirectory(directory + '/' + dir_name1)) and not(isExistDirectory(directory + '/' + dir_name2)):
        os.chdir(directory)
        
        try:
            os.system("mkdir {0} && cd {0} && mkdir {1} && mkdir {2}".format(dir_name1, dir_name11, dir_name12))
            os.chdir(directory)
            os.system("mkdir {0} && cd {0} && mkdir {1} && mkdir {2}".format(dir_name2, dir_name21, dir_name22))
        except:
            printError("Un probleme est survenu lors de la creation des repertoires.")
            exit()
        else:
            printSuccess("Les repertoires ont ete crees avec succes")
            os.chdir(directory)
            os.system("tree {} && tree {}".format(dir_name1, dir_name2))
            return True

    else:
        print("Repertoire existant!")
        yes_or_no = input("Voulez-le sauvegarder dans ~/Sauvegarde ?? o/n > ")
        if yes_or_no == 'o' or yes_or_no == 'O':

            if isExistDirectory(directory + '/Sauvegarde'):
                try:
                    os.system("mkdir {0} && cd {0} && mkdir {1} && mkdir {2}".format(dir_name1, dir_name11, dir_name12))
                    os.chdir(directory)
                    os.system("mkdir {0} && cd {0} && mkdir {1} && mkdir {2}".format(dir_name2, dir_name21, dir_name22))
                except:
                    printError("Un probleme est survenu lors de la creation des repertoires.")
                    exit()
                else:
                    printSuccess("Les repertoires ont ete crees avec succes")
                    os.chdir(directory)
                    os.system("tree Sauvegarde")
                    return True
            
            else:
                os.chdir(directory)
                try:
                    os.system("mkdir Sauvegarde && cd Sauvegarde && mkdir {0} && cd {0} && mkdir {1} && mkdir {2}".format(dir_name1, dir_name11, dir_name12))
                    os.chdir(directory + "/Sauvegarde")
                    os.system("mkdir {0} && cd {0} && mkdir {1} && mkdir {2}".format(dir_name2, dir_name21, dir_name22))
                except:
                    printError("Un probleme est survenu lors de la creation des repertoires.")
                    exit()
                else:
                    printSuccess("Les repertoires ont ete crees avec succes")
                    return True

        else:
            printSuccess("Fin du programme")
            exit()

dir_exist = isExistDirectory(directory)

if dir_exist == False:
    exit()

while True:
    print("")
    print("0- Creer un repertoire")
    print("1- Creer 02 repertoires")
    print("2- Creer un fichier")
    print("3- Type de fichiers et permissions")
    print("4- Information CPU & memoire et etat des disques et partitions du disque")
    print("5- Type de disque de l'ordinateur")
    print("6- Espace disque et memoire disponible")
    print("7- M-a-j des referentiels avec un depot PPA")
    print("8- Nombre et liste d'utilisateurs presents dans le systeme")
    print("9- Faire expirer le mot de passe d'un compte utilisateur apres 5h")
    print("10- Creer et ajouter un utilisateur")
    print("11- Installer le service DHCP")
    print("12- Afficher les parametres reseaux")
    print("Q/q- QUITTER")
    print("\n\nSélectionner l’option de votre choix : ")
    var1 = input("\n----->")
    os.system("clear")
    try:
        var = int(var1)
    except:
        var = ''
    """
    if var != 0 and count == 0:
        print("\nVous devez forcement debuter par 0 ")
        continue
    else:
        count += 1
    
    if nb0 >= 2 and var == 0:
        print("\nVous ne pouver plus choisir 0 ")
        continue
    """

    if var == 0:
        nb0 += 1
        printSuccess("0- Creer un repertoire")
        dir_name = input("Entrez le nom du repertoire: ")

        if isExistDirectory(directory + '/' + dir_name):
            os.chdir(directory)
            yes_or_no = input("Repertoire existant. le sauvegarder dans {} ? o/n : > ".format(directory + '/Sauvegarde'))
        
            if yes_or_no == 'o' or yes_or_no == 'O':
                if isExistDirectory(directory + '/Sauvegarde'):
                    try:
                        os.system("mkdir {}".format(dir_name))
                    except:
                        printError("Echec de creation du repertoire : {} \n\tFin du programme".format(dir_name))
                        exit()
                    else:
                        printSuccess("Repertoire cree avec succes")
                else:
                    try:
                        os.system("mkdir Sauvegarde && cd Sauvegarde && mkdir {}".format(dir_name))
                    except:
                        printError("Echec de creation du repertoire : {} \n\tFin du programme".format(dir_name))
                        exit()
                    else:
                        printSuccess("Repertoire cree avec succes")

            else:
                printSuccess("Fin du programe")
                exit()
        else:
            os.chdir(directory)
            try:
                os.system("mkdir {}".format(dir_name))
            except:
                printError("Echec de creation du repertoire : {} \n\tFin du programme".format(dir_name))
                exit()
            else:
                printSuccess("Repertoire cree avec succes")

    elif var == 1:
        printSuccess("1- Creer 02 repertoires")
        question1()

    elif var == 2:
        printSuccess("2- Creer un fichier")

        file_name1 = input("Entrez le nom du fichier1: ")
        file_name11 = file_name1 + '.1'
        file_name12 = file_name1 + '.2'

        file_name2 = input("Entrez le nom du repertoire2: ")
        file_name21 = file_name2 + '.1'
        file_name22 = file_name2 + '.2'

        """
            avant de creer les fichier, tester dabord si ils existe ou pas>>> 
            
        """
        if not(isExistDirectory(directory + '/' + dir_name1 + '/' + dir_name11)) or not(isExistDirectory(directory + '/' + dir_name1 + '/' + dir_name12)) or not(isExistDirectory(directory + '/' + dir_name2 + '/' + dir_name21)) or not(isExistDirectory(directory + '/' + dir_name2 + '/' + dir_name22)):
            result = question1()
            while result == False:
                result = question1()

        os.chdir(directory)
        
        try:
            os.system("cd {0} && cd {1} && touch {2}".format(dir_name1, dir_name11, file_name11))
            os.system("cd {0} && cd {1} && touch {2}".format(dir_name1, dir_name11, file_name11))
            os.system("cd {0} && cd {1} && touch {2}".format(dir_name1, dir_name12, file_name12))
            os.system("cd {0} && cd {1} && touch {2}".format(dir_name2, dir_name21, file_name21))
            os.system("cd {0} && cd {1} && touch {2}".format(dir_name2, dir_name22, file_name22))
        except:
            printError("Un probleme est survenu lors de la creation des repertoires.")
            exit()
        else:
            printSuccess("Les repertoires ont ete crees avec succes")

    elif var == 3:
        printSuccess("3- Type de fichiers et permissions")
        if dir_name1 == '' or dir_name2 == '':
            question1()

        rep11 = dir_name1 + '/' + dir_name11 
        rep12 = dir_name1 + '/' + dir_name12
        rep21 = dir_name2 + '/' + dir_name21 
        rep22 = dir_name2 + '/' + dir_name22 

        try:
            os.system("file {0} && ls -l {0} && file {1} && ls -l {1}".format(rep11, rep12))
            os.system("file {0} && ls -l {0} && file {1} && ls -l {1}".format(rep21, rep22))
        except:
            print("impossible d'afficher les informations...")
   
    elif var == 4:
        printSuccess("4- Information CPU & memoire et etat des disques et partitions du disque")
        try:
            os.system("lscpu && cat /proc/cpuinfo && df")
        except:
            printError("Une erreur s'est produite lors de l'execution de la commade")
            exit()
        
    elif var == 5:
        printSuccess("5- Type de disque de l'ordinateur")
        try:
            os.system("lsblk -o name,type")
        except:
            printError("Erreur lors de l'execution de la commade")

    elif var == 6:
        printSuccess("6- Espace disque et memoire disaponible")
        try:
            os.system("df -h; free -h")
        except:
            printError("Erreur survenu lors de l'executionde la commande")


    elif var == 7:
        printSuccess("7- M-a-j des referentiels avec un depot PPA")
        nomDepot = input("Entrez le nom du packet ")
        try:
            os.system("sudo add-apt-repository ppa:{} && sudo ppa upgrade".format(nomDepot)) 
        except:
            printError("Erreur survenu lors de l'executionde la commande")


    elif var == 8: 
        printSuccess("8- Nombre et liste d'utilisateurs presents dans le systeme")
        try:
            os.system("cut -d: -f1 /etc/passwd")
        except:
            printError("Erreur survenu lors de l'executionde la commande")
            exit()

    elif var == 9:
        printSuccess("9- Faire expirer le mot de passe d'un compte utilisateur apres 5h")
        usr_name = input("Nom de l'utilisateur: > ")
        try:
            os.system("echo usermod --lock {} | at now + 300 minutes".format(usr_name))
        except:
            printError("Erreur survenu lors de l'executionde la commande")

    elif var == 10:
        printSuccess("10- Creer et ajouter un utilisateur")
        usr_name = input("Nom de l'utilisateur: > ")
        try:
            os.system("sudo adduser {}".format(usr_name))
        except:
            printError("Erreur survenu lors de l'executionde la commande")
 
    elif var == 11: 
        printSuccess("11- Installer le service DHCP")
        try:
            os.system("sudo apt install isc-dhcp-server -y")
        except:
            printError("Erreur survenu lors de l'executionde la commande")
 
    elif var == 12: 
        printSuccess("12- Afficher les parametres reseaux")
        url = input("Entrez l'url : ")
        try:
            os.system("ifconfig; wget {0}".format(url))
        except:
            printError("Erreur survenu lors de l'executionde la commande")

    elif var1 == 'q' or var1 == 'Q':
        printSuccess("Fin du Programme")
        exit()
```
